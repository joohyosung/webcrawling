```python
class Char:
    def __init__(self,name,attack,hp):
        self.name = name
        self.attack = attack
        self.hp = hp
    def info(self):
        print(f'''
              이름 : {self.name}\n 
              공격 : {self.attack}\n
              체력 : {self.hp}
              ''')
        
        
lion = Char('','100','200')
lion.info()
    
        
```
